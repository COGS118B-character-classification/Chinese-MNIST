```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#code to read the data into one dataframe
df = pd.DataFrame()
for i in range(5):
    df_temp = pd.read_csv(f'../data/processed_chinese_mnist_part_{i}.csv')
    df = df.append(df_temp)
df = df.reset_index(drop = True)
```

```python
df.cn_label.value_counts()
```

```python
zero_arrs = df[df.cn_label == '零']
num = 0 
fig, axs = plt.subplots(10, 10, figsize = (20, 20))
for i in range(10):
    for j in range(10):
        axs[i, j].imshow(zero_arrs.iloc[num][3:].values.astype(float).reshape(64, 64), cmap = 'gray')
        num += 1
```

```python
arrs = df[df.cn_label == '一']
num = 0 
fig, axs = plt.subplots(10, 10, figsize = (20, 20))
for i in range(10):
    for j in range(10):
        axs[i, j].imshow(arrs.iloc[num][3:].values.astype(float).reshape(64, 64), cmap = 'gray')
        num += 1
```
