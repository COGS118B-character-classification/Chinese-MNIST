```python
zero_arrs = df[df.cn_label == '零']
num = 0 
fig, axs = plt.subplots(10, 10, figsize = (20, 20))
for i in range(10):
    for j in range(10):
        axs[i, j].imshow(zero_arrs.iloc[num][3:].values.astype(int).reshape(64, 64), cmap = 'gray')
        num += 1
```

```python
arrs = df[df.cn_label == '一']
num = 0 
fig, axs = plt.subplots(10, 10, figsize = (20, 20))
for i in range(10):
    for j in range(10):
        axs[i, j].imshow(arrs.iloc[num][3:].values.astype(int).reshape(64, 64), cmap = 'gray')
        num += 1
```
